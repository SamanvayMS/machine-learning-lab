```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.model_selection import GridSearchCV, StratifiedKFold, cross_validate
```

```python
import time
```

```python
parameters={'n_estimators':(50,100,200,300,400,500)}
forest=RandomForestClassifier(oob_score=True)
kf = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)
model=GridSearchCV(forest,parameters,cv=kf)
```

```python
start_time = time.time()
model.fit(df.drop(['DEFAULT','ID'],axis=1),df['DEFAULT'])
end_time = time.time()
total_time=int(end_time-start_time)
print(f'The grid search CV took {total_time:.2f}s to run')
```

# looping over the values instead of using a grid search CV to get train and test values 

```python
trees = [50,100,200,300,400,500]
```

```python
acc_scores=[]
val_scores=[]
calc_time=[]
for tree in trees:
    start_time = time.time()
    forest1=RandomForestClassifier(n_estimators=tree,oob_score=True)
    scores = cross_validate(forest1,df.drop(['DEFAULT','ID'],axis=1),df['DEFAULT'], cv=kf, return_train_score=True)
    end_time = time.time()
    test_acc = np.mean(scores['test_score'])
    train_acc = np.mean(scores['train_score'])
    test_std = np.std(scores['test_score'])
    train_std = np.std(scores['train_score'])
    print(f'{train_acc:.5f} , {test_acc:.5f}')
    acc_scores.append(train_acc)
    val_scores.append(test_acc)
    calc_time.append(end_time-start_time)
```

```python
for i in range(len(trees)):
    print(f' for {trees[i]} trees/n_estimators took {calc_time[i]:.2f}s , in sample accuracy :- {acc_scores[i]:.5f} , out of sample accuracy:- {val_scores[i]:.5f}')
```

```python
plt.plot(trees,calc_time)
plt.title('number of trees vs time')
plt.xlabel('number of trees')
plt.ylabel('time in s')
plt.show()
```

```python
total_time=np.sum(calc_time)
```

```python
total_time
```

Here we compare and see the time taken by a grid search CV and a for loop. the advantage of using a for loop is we can decide the optimum number of trees not just based on the score but also on the time it takes to run as there's not much rise in accuracy and in some instances speed matters. we see that the compuation time increases with increasing number of trees almost linearly.

Feature importance is a measure of the contribution of each feature in a machine learning model to the performance of the model. It is used to identify the most important features in the dataset and to gain insight into the underlying relationships between the features and the target variable.Random forests and other tree based models can provide feature importance scores based on the decrease in impurity (e.g., Gini impurity or entropy) caused by each feature when splitting the data. Features with higher importance scores are considered more important.
