```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.model_selection import StratifiedKFold, cross_validate
```

```python
clf = DecisionTreeClassifier()
```

```python
def random_split_tree(seed):
    X_train, X_test, y_train, y_test = train_test_split(df.drop(['DEFAULT','ID'],axis=1),df['DEFAULT'], test_size=0.1, random_state=seed, stratify=df['DEFAULT'])
    clf.fit(X_train, y_train)
    out_sample_accuracy = accuracy_score(y_test, clf.predict(X_test))
    in_sample_accuracy = accuracy_score(y_train, clf.predict(X_train))
    return out_sample_accuracy,in_sample_accuracy
```

```python
import time
```

```python
start_time = time.time()

out_sample=np.zeros(10)
in_sample=np.zeros(10)
for seed in range(0,10):
    out_sample[seed],in_sample[seed] = random_split_tree(seed+1)
    
end_time = time.time()

print(f'time taken = {end_time - start_time:.2f}')

out_sample_accuracy = np.mean(out_sample)
in_sample_accuracy = np.mean(in_sample)
out_sample_std = np.std(out_sample)
in_sample_std = np.std(in_sample)
```

```python
print(f'out_sample_accuracy:-{out_sample_accuracy:.5f} ,in_sample_accuracy:- {in_sample_accuracy:.5f} ,out_sample_std:-{out_sample_std:.5f} , in_sample_std:-{in_sample_std:.5f}')
```

```python
kf = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)
```

```python
start_time = time.time()

scores = cross_validate(clf,df.drop(['DEFAULT','ID'],axis=1),df['DEFAULT'], cv=kf, return_train_score=True)

end_time = time.time()

print(f'time taken = {end_time - start_time:.2f}')

test_acc = np.mean(scores['test_score'])
train_acc = np.mean(scores['train_score'])
test_std = np.std(scores['test_score'])
train_std = np.std(scores['train_score'])
```

```python
print(f'test_acc:-{test_acc:.5f} , train_acc:-{train_acc:.5f} ,test_std:-{test_std:.5f} , train_std:-{train_std:.5f}')
```

Kfold gives us very similar results but the time for execution is faster and is more efficient and standard deviation of values appears to be better hence would be a better choice. here i have used stratified K fold as the labels were not uniform and there was too many current values when compared to default.
